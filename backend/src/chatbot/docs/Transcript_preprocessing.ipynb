````python
doc_name = "Communication Section Video Transcript.docx"
````

````python
from docx import Document
from fpdf import FPDF

# Load the Word document
doc = Document(doc_name)

# Initialize variables
videos = []
current_video = []
current_title = None
current_url = None

# Loop through each paragraph in the document
for para in doc.paragraphs:
    text = para.text.strip()
    print(f"Processing line: {text}")  # Debug print

    # Detect title line (assuming it starts with #)
    if text.startswith("# ") and "http" not in text:
        if current_video:
#             print(f"Appending video: {current_title}")  # Debug print
            videos.append({
                'title': current_title,
                'url': current_url,
                'transcript': " ".join(current_video).replace("  ", " ").strip()
            })
            current_video = []
        current_title = text[2:].strip()  # Remove the "# " and strip whitespace
        current_url = None  # Reset URL
        print(f"New title detected: {current_title}")  # Debug print
    elif text.startswith("# http"):
        current_url = text[2:].strip()  # Save URL (removing the "# ")
        print(f"URL detected: {current_url}")  # Debug print
    else:
        # Filter out empty lines and timestamps
        if text.startswith("00:"):
            current_video.append(text[13:])  # Add the text line
#             print(f"Added line to transcript: ```{text}```")  # Debug print

# Don't forget to add the last video
if current_video:
    print(f"Appending last video: {current_title}")  # Debug print
    videos.append({
        'title': current_title,
        'url': current_url,
        'transcript': " ".join(current_video).replace("  ", " ").strip()
    })

# Check the content of the videos array
print(f"Total videos parsed: {len(videos)}")
for video in videos:
    print(f"Video Title: {video['title']}, URL: {video['url']}, Transcript: {video['transcript'][:50]}...")  # Print first 50 characters of transcript

# Function to save a video transcript to a PDF file with UTF-8 encoding
def save_to_pdf(video, index):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    # Ensure the text is encoded in UTF-8
    pdf.set_font("Arial", 'B', size=14)
    pdf.multi_cell(0, 10, video['title'].encode('latin-1', 'replace').decode('latin-1'))

    pdf.set_font("Arial", 'I', size=12)
    pdf.multi_cell(0, 10, video['url'].encode('latin-1', 'replace').decode('latin-1'))
    
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, video['transcript'].encode('latin-1', 'replace').decode('latin-1'))

    # Save with a filename based on the title, but sanitize it to avoid filesystem issues
    sanitized_title = ''.join(c if c.isalnum() else '_' for c in video['title'])
    pdf.output(f"{sanitized_title}.pdf")
    
def save_to_word(video, index):
    doc = Document()

    # Add title
    doc.add_heading(video['title'], level=1)

    # Add URL
    doc.add_paragraph(video['url'], style='Intense Quote')

    # Add transcript
    doc.add_paragraph(video['transcript'])

    # Save with a filename based on the title, but sanitize it to avoid filesystem issues
    sanitized_title = ''.join(c if c.isalnum() else '_' for c in video['title'])
    doc.save(f"{sanitized_title}.docx")

# Loop through each video and save it as a separate PDF
for i, video in enumerate(videos):
    save_to_word(video, i)

print("PDF files created successfully.")
````
